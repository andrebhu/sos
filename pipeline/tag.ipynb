```python
import pandas as pd
import requests
```

```python
df = pd.read_csv("000000000000")
```

```python
df["parent_id"]
```

```python
r = None
all_tags = []
tags = []
for i in range(0, len(df), 30):
  r = df.iloc[i:i+30]["parent_id"]
  vals = r.tolist()
  vals = [str(v) for v in vals]

  ids = "%3B".join(vals)
  url = f"https://api.stackexchange.com/2.3/questions/{ids}?&order=desc&sort=activity&site=stackoverflow"
  response = requests.get(url)

  try:
    for i in response.json()["items"]:
      tags.append(i["tags"])
  except:
    break
  

tags
len(tags)

```
